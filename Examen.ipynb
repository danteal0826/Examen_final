#Mapa coropletico

```python
import pandas as pd
```

```python
municipios= pd.read_csv('/content/xn--Poblacin_2017_en_Cundinamarca-12c.csv')
municipios
```

```python
df_municipios=municipios.reset_index()
df_municipios
```

```python
df = df_municipios[['f2', 'women']]
df
```

```python
import geopandas as gpd
shapefile_path = '/content/xn--Poblacin_2017_en_Cundinamarca-12c (1).csv'
gdf_municipios = gpd.read_file(shapefile_path)
gdf_municipios = gdf_municipios.merge(df, left_on='f2', right_on='f2', how='left')
gdf = gdf_municipios[['f2', 'women']]
gdf
```

```python
import matplotlib.pyplot as plt
mapa=gdf.plot(
    column='women',
    figsize=(15,15),
    edgecolor="black",
    cmap='tab20c',
    legend=True,
    legend_kwds={
        'loc':'upper center',
        'bbox_to_anchor': (0.5, -0.05),
        'ncol':8
    }
)
plt.title("Mapa localidad Engativá", fontsize=50)
plt.savefig("mapa.pdf", bbox_inches='tight')
```
